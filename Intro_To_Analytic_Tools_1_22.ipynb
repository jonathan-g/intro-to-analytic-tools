## Basic Plotting

One of the first steps for most types of data analysis is to plot some of the data to explore it and get an intuition for the types of analysis that might work. We will work with two types of plots for this dataset: scatter and histogram.

### Scatter Plot
Essentially the most basic of plot is the scatter plot. It simply plots one small dot, or point, per data sample. In the case of the iris data, this corresponds to plotting one point for each flower that we have measurements for. But before we plot the data, we have to decide what attributes we want to plot. A basic graph has 2 axis, x and y. So, we can choose two of the four features of the iris data that we want to plot. Lets start by choosing sepal length for x and sepal width for y. We can use the bracket operator to extract the desired columns from the pandas array.



```
x = iris_data['sepal_length']
y = iris_data['sepal_width']
```

Now that we have our data to use for each axis, matplotlib makes it really easy to construct a scatter plot by simply calling the *scatter()* function. This function takes in a minimum of two parameters (or sometimes called arguments). The first is the data for the x axis and the second is the data for the y. Finally, after we call scatter, we can call the *show()* function to display the plot here. Lets generate a simple plot using our data.

```
plt.scatter(x, y)
plt.show()
```

### Histogram

This can provide valuable insights about how your data is distributed and the kinds of categories that can be formed. ## Probability Distributions

When we look at data, we often want to tell how the data is distributed, which can help give us a sense of what to expect from the data and what statistical tests we might use. There are several common statistical distributions which you should be familiar with.

### Uniform Distribution
First, we have a uniform distribution. In this distribution, every possible data value is equally likely to occur. Lets take a look at what a uniform distribution looks like. We can use Scipy to generate a collection of uniformly distributed data points. When we plot these, you can see that they are approximately even throughout the ranage of values in which they were generated.

```
from scipy.stats import uniform

n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)

plt.scatter(list(range(n)), data_uniform, s=3)
plt.title("Uniform Distribution Scatter Plot")
plt.show()
```

When we are analyzing distributions, in addition to a scatter plot it is often helpful to plot a histogram, which approximated the probability density function of the data. When we plot a histogram of the uniform data, you can see that it looks like a nearly flat line across the axis. This is because at any point in the range, the probability is exactly equal.

```
plt.hist(data_uniform, bins=15)
plt.title("Histogram to Approximate A Uniform Distribution PDF")
plt.show()
```

### Normal Distribution

A normal (or guassian distribution) is one of the most common and well understood types of distributions. Instead of the probability being equal across the range of data, it instead has increasing probability as you approch the center of the range. Lets scatter plot a normal distribution to see how it differs from a uniform distribution.

```
from scipy.stats import norm

data_normal = norm.rvs(size=10000,loc=0,scale=1)

plt.scatter(list(range(10000)), data_normal, s=3)
plt.title("Normal Distribution Scatter Plot")
plt.show()
```

So you can see from this plot, the data is much more dense in the center (the mean) than it is at larger and smaller values. This is generally the idea of a normal distribution and appears very often in natural data. Lets plot a histogram of the normal distribution to draw an estimate of the probability density function.

```
plt.hist(data_normal)
plt.title("Histogram to Approximate a Normal Distribution PDF")
plt.show()
```

You can see that instead of being a stright line all the way across, there is instead a curve leading to the mean of the data.
